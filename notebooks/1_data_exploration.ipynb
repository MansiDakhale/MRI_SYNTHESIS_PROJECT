Loading & Visualizing a NIFTI File


```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# Sample file path (update this with an actual file)
sample_nifti_path = r"C:\Users\OMEN\Desktop\MRI_Synthesis_Project\MRI_SYNTHESIS_PROJECT\data\Masks\test_masks\ISPY1_1002\52000_pe_segmentation_thresh70_mask.nii.gz" 

# Load the NIfTI file
nifti_img = nib.load(sample_nifti_path)
data = nifti_img.get_fdata()

# Print shape
print(f"Image Shape: {data.shape}")  # Expected: (H, W, D)

# Show a few slices
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for i, slice_idx in enumerate([data.shape[2]//4, data.shape[2]//2, 3*data.shape[2]//4]):
    axes[i].imshow(data[:, :, slice_idx], cmap="gray")
    axes[i].set_title(f"Slice {slice_idx}")
    axes[i].axis("off")
plt.show()
```

Compare ceT1 and preT1 Modalities

```python
# Load both modalities for the same patient
preT1_path = r"C:\Users\OMEN\Desktop\MRI_Synthesis_Project\MRI_SYNTHESIS_PROJECT\data\nifti\ISPY1_1015\PreT1\1_t1-axial-locator.nii.gz"
ceT1_path = r"C:\Users\OMEN\Desktop\MRI_Synthesis_Project\MRI_SYNTHESIS_PROJECT\data\nifti\ISPY1_1015\CeT1\41001_dynamic-3dfgre_pe1.nii.gz"

preT1 = nib.load(preT1_path).get_fdata()
ceT1 = nib.load(ceT1_path).get_fdata()

# Display side by side
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(preT1[:, :, preT1.shape[2]//2], cmap="gray")
ax[0].set_title("Pre-Contrast (T1)")
ax[1].imshow(ceT1[:, :, ceT1.shape[2]//2], cmap="gray")
ax[1].set_title("Post-Contrast (ceT1)")
plt.show()
```

Check Segmentation Masks

```python
seg_path = r"C:\Users\OMEN\Desktop\MRI_Synthesis_Project\MRI_SYNTHESIS_PROJECT\data\nifti\ISPY1_1001\Segmentation\32001_breast_tissue_segmentation.nii.gz"

seg = nib.load(seg_path).get_fdata()

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(ceT1[:, :, ceT1.shape[2]//2], cmap="gray")
ax[0].set_title("MRI Image (ceT1)")
ax[1].imshow(seg[:, :, seg.shape[2]//2], cmap="hot", alpha=0.7)
ax[1].set_title("Segmentation Mask Overlay")
plt.show()
```
